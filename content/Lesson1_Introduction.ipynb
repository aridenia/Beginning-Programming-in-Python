# Introduction to Interactive Python (IPython)
Welcome to your first interactive lesson running entirely in the browser with JupyterLite.


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
print("Hello from JupyterLite! Python is running directly in your browser.")
```

```python
@widgets.interact(x=(0, 20))
def square(x):
    return f"{x} squared is {x**2}"
```

```python
@widgets.interact(freq=(1, 10), amplitude=(1, 5))
def plot_wave(freq=1, amplitude=1):
    x = np.linspace(0, 2*np.pi, 400)
    y = amplitude * np.sin(freq * x)

    plt.figure(figsize=(6, 3))
    plt.plot(x, y)
    plt.title(f"Sine Wave: freq={freq}, amplitude={amplitude}")
    plt.grid(True)
    plt.show()
```

```python
@widgets.interact(freq=(1, 10))
def plot_cos(freq=1):
    x = np.linspace(0, 2*np.pi, 400)
    y = np.cos(freq * x)

    plt.figure(figsize=(6, 3))
    plt.plot(x, y, color="orange")
    plt.title(f"Cosine Wave: freq={freq}")
    plt.grid(True)
    plt.show()
```
