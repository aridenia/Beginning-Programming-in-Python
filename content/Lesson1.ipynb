```python
import numpy as np
import matplotlib.pyplot as plt

print("Hello from JupyterLite! Python is running directly in your browser.")
```

```python
def square(x):
    return f"{x} squared is {x**2}"

# Simple text input for interactivity
x = int(input("Choose a number between 0 and 20: "))
square(x)
```

```python
def plot_wave(freq=1, amplitude=1):
    x = np.linspace(0, 2*np.pi, 400)
    y = amplitude * np.sin(freq * x)

    plt.figure(figsize=(6, 3))
    plt.plot(x, y)
    plt.title(f"Sine Wave: freq={freq}, amplitude={amplitude}")
    plt.grid(True)
    plt.show()

# Simple loop to simulate interactivity
while True:
    print("\n--- Sine Wave Plot ---")
    freq = int(input("Frequency (1–10): "))
    amp = int(input("Amplitude (1–5): "))
    plot_wave(freq, amp)

    cont = input("Plot another? (y/n): ").lower()
    if cont != "y":
        break
```

```python
def plot_cos(freq=1):
    x = np.linspace(0, 2*np.pi, 400)
    y = np.cos(freq * x)

    plt.figure(figsize=(6, 3))
    plt.plot(x, y, color="orange")
    plt.title(f"Cosine Wave: freq={freq}")
    plt.grid(True)
    plt.show()

# Simulated interactivity
while True:
    print("\n--- Cosine Wave Plot ---")
    freq = int(input("Frequency (1–10): "))
    plot_cos(freq)

    cont = input("Plot another? (y/n): ").lower()
    if cont != "y":
        break
```
